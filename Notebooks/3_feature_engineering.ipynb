# Feature Engineering

### 1. Transformation
    
- EDA revealed that many features were skewed (i.e. not normally distributed). Here we will use scipy.stats to figure out which features are skewed and apply a log transform to reduce the skew. There are other methods to deal with skew, but we will stick to this simple method for now. 

- Applying the log transform did help reduce the skew, which is likely to improve performance. 

- Of note, based on our criteria, the target variable was not transformed, so we will not have to apply an inverse log pprior to making predictions.
    
    
    
### 2. Encoding

- We have one categorical feature ('Status') which we will encode

### 3. Feature selection

- EDA reveled that some features were strongly correlated to each other. We may later choose to drop these features to improve performance. For now we will leave them in. 

```python
def find_skew(dataset, show=True, col_names=False):
  df_skew = pd.DataFrame(dataset.select_dtypes(np.number).columns, columns=['Feature'])
  df_skew['Skew'] = df_skew['Feature'].apply(lambda feature: scipy.stats.skew(dataset[feature]))
  df_skew['Absolute Skew'] = df_skew['Skew'].apply(abs)
  df_skew['Skewed'] = df_skew['Absolute Skew'].apply(lambda x: True if x >= 0.65 else False)

  if show:
    from IPython.display import display
    display(df_skew)

  if col_names:
    return df_skew[df_skew['Absolute Skew'] >= 0.65]['Feature'].values

find_skew(df_train)
```

```python
df = df_train.copy()
skewed_cols = find_skew(df, show=False, col_names=True)

for col in skewed_cols:
  df[col] = np.log1p(df[col])

find_skew(df)
```

# Create EngineerFeatures Object

```python
class EngineerFeatures():

    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None): 
        X_new = X.copy()

        # Drop uncorrelated or unimportant features
        cols_to_drop = [
            'Country', # This column has been encoded and imputation is done. It can now be dropped
            'Population' # Very weak correaltion to target variable
        ]
        
        X_new = X_new.drop(cols_to_drop, axis=1)

        # Feature transformations for numeric features
        # Note that skewed columns will have to be entered manually here 
        skewed_cols = [
            'Adult Mortality',
            'infant deaths', 
            'percentage expenditure', 
            'Hepatitis B',
            'Measles ', 
            'under-five deaths ',
            'Polio',
            'Diphtheria ', 
            ' HIV/AIDS', 
            'GDP',
            ' thinness  1-19 years', 
            ' thinness 5-9 years',
            'Income composition of resources'
        ]

        for col in skewed_cols:
            X_new[col] = np.log1p(X_new[col])

        # Feature encoding
        cols_to_encode = [
            'Status',
        ]
        
        X_new = pd.get_dummies(X_new, columns=cols_to_encode, drop_first=True)

        return X_new

    def fit_transform(self, X, y=None):
        return self.fit(X).transform(X)
```

```python
engg = EngineerFeatures()

df_train_engineered = engg.fit_transform(df_train)
df_test_engineered = engg.transform(df_test)
```

```python
# Note that the targeet variable is unchanged, so we can proceed with saving just the features
X_train_engineered = df_train_engineered.drop('Life expectancy ', axis=1)
X_test_engineered = df_test_engineered.drop('Life expectancy ', axis=1)
```
